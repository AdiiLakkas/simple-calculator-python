```python
import math
num1 = input("Enter the first number :")                                             #Take first number from user
operator = input("Enter the operator you wish to perfrom (+,-,*,/) :")               #Take operator from user
num2 = input("Enter the second number :")                                            #Take second number from user

#convert input from string to float
num1 = float(num1)
num2 = float(num2)

#check which operator the user entered
if operator == '+':
    result = num1 + num2     #Addition
elif operator == '-':
    result = num1-num2      #Substraction
elif operator == '*':
    result = num1*num2      #Multiplication
elif operator == '/':
    if num2 == 0:
        print("Error ! Division by zero is not allowed")
    else:
        result = num1/num2            #Division
else:
    print("Entered invalid operator!! please check..!!")
    
print(num1,operator,num2, '=' , result)
```
